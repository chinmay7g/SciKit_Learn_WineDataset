```python
#importing essentials

import numpy as np
from sklearn.datasets import load_wine
```

```python
#retreiving data
data = load_wine()
data['DESCR']
```

```python
X = data['data']
y = data['target']
```

```python
#shape and other information

print(X.shape)
```

```python
#importing estimators

from sklearn.base import BaseEstimator, ClassifierMixin
```

```python
#learning about counter
from collections import Counter

#printing y
y
```

```python
c = Counter(y)
c

c.most_common(1)[0][0]
```

```python
#defining and building a class

class MajorityClassifier(BaseEstimator, ClassifierMixin):
    def __init__(self):
        pass
    
    def fit(self,X,y):
        c = Counter(y)
        self.mode_ = c.most_common(1)[0][0]
        self.fraction = np.array(list(c.values())) / X.shape[0]
        return self
    
    def predict(self,X):
        return self.mode_ * np.ones(X.shape[0])
        
    
    def predict_probability(self,X):
        return np.dot(np.ones((X.shape[0],1)), self.fraction.reshape(1,3))
        
    
```

```python
mC = MajorityClassifier()
```

```python
mC.fit(X,y)
mC.predict(X)
```

```python
mC.predict_probability(X)
```
